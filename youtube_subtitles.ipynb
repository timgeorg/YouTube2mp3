```python
def split_text_from_test(test, max_length=16000):
    current_length = 0
    current_chunk = []
    chunks = []
    timestamps = []

    # Idea:
    # Create a list of dict first, get end time and text content
    # Then split into chunks that match the chapter of the video

    for item in test:
        word = item['text']
        end_time = int(item['start']) + int(item['duration'])
        minutes, seconds = divmod(end_time, 60)
        end_time_formatted = f"{minutes}m {seconds}s"
        if current_length + len(word) + 1 > max_length:
            chunks.append(" ".join(current_chunk))
            current_chunk = [f'{word} ({end_time_formatted})']
            current_length = len(word) + 1
        else:
            current_chunk.append(f'{word} ({end_time_formatted})')
            timestamps.append(end_time)
            current_length += len(word) + 1

    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks

split_texts = split_text_from_test(test)
split_texts
```

```python
# importing the libraries 
from bs4 import BeautifulSoup 
import requests 
  
# creating function 
def scrape_info(url): 
      
    # getting the request from url 
    r = requests.get(url) 
      
    # converting the text 
    s = BeautifulSoup(r.text, "html.parser") 
      
    # finding the button and clicking it
    button = s.find(id ='description')
    if button:
        button.click()
        # finding the description
        description = s.find('ytd-text-inline-expander', {'id': 'description-inline-expander'})
        if description:
            print(description.text.strip())
```

```python

# URL of the video 
url ="https://www.youtube.com/watch?v=W_JfzXaYNDI"
      
# calling the function 
scrape_info(url) 
     
```
